```python
import pandas as pd
import numpy as np
import pylab as pl
from sklearn import tree
from sklearn import preprocessing

pd.set_option('display.max_rows',100)
pd.set_option('display.max_columns',60)
```

```python
#bring in the bank marketing data, set ; as delimiter
df = pd.read_csv("bank.csv", delimiter=';')

```

```python
#have a look at the data
df.info()
```

```python
df.head()
```

```python
#define the LabelEncoder to change the features data to something ML friendly
le = preprocessing.LabelEncoder()
```

```python
dfle = df
dfle.job = le.fit_transform(dfle.job)
dfle.marital = le.fit_transform(dfle.marital)
dfle.education = le.fit_transform(dfle.education)
dfle.default = le.fit_transform(dfle.default)
dfle.housing = le.fit_transform(dfle.housing)
dfle.loan = le.fit_transform(dfle.loan)
dfle.contact = le.fit_transform(dfle.contact)
dfle.month = le.fit_transform(dfle.month)
dfle.poutcome = le.fit_transform(dfle.poutcome)
dfle.y = le.fit_transform(dfle.y)
dfle.head()
```

```python
#define the feature set
features = dfle.ix[:, 0:16]
```

```python
model = tree.DecisionTreeClassifier(random_state=0)
```

```python
#run the decision tree model
model.fit(features, dfle.y)
```

```python
#have a look at accuracy
pd.crosstab(dfle.y, model.predict(features))
```

```python
#bring in cross-validation
from sklearn.cross_validation import KFold

def cross_validate(X, y, classifier, k_fold) :

    # derive a set of (random) training and testing indices
    k_fold_indices = KFold(len(X), n_folds=k_fold,
                           shuffle=True, random_state=0)

    k_score_total = 0
    # for each training and testing slices run the classifier, and score the results
    for train_slice, test_slice in k_fold_indices :

        model = classifier(X[ train_slice  ],
                         y[ train_slice  ])

        k_score = model.score(X[ test_slice ],
                              y[ test_slice ])

        k_score_total += k_score

    # return the average accuracy
    return k_score_total/k_fold
```

```python
model = tree.DecisionTreeClassifier(random_state=0).fit
target = dfle['y'].values
features = features.values
cross_validate(features, target,model,10)
```
